Changing all the strings to numbers if possible:

```python
df['gender'] = df['gender'].replace({'Male': 0, 'Female': 1})
"""
Note:
We used the binary datatype where:
Male = 0
Female = 1
"""
# Display the unique values and their counts to verify the transformation
display(df['gender'].value_counts())
```

```python
df['physical_activity_level'] = df['physical_activity_level'].replace({'Low': 0, 'Medium': 1, 'High': 2})
"""
Note:
We used the ordinal datatype where:
Low = 0
Medium = 1
High = 2
"""
# Display the unique values and their counts to verify the transformation
display(df['physical_activity_level'].value_counts())
```

```python
df['city_type'] = df['city_type'].replace({'Rural': 0, 'Semi-Urban': 1, 'Urban': 2})
"""
Note:
We used the ordinal datatype where:
Rural = 0
Semi-Urban = 1
Urban = 2
"""
# Display the unique values and their counts to verify the transformation
display(df['city_type'].value_counts())
```

```python
# Fill missing 'insurance_type' values with the mode
mode_insurance_type = df['insurance_type'].mode()[0]
df['insurance_type'].fillna(mode_insurance_type, inplace=True)

# Display the updated value counts to confirm
display(df['insurance_type'].value_counts(dropna=False))
```

```python
df['insurance_type'] = df['insurance_type'].replace({'Government': 0, 'Private': 1})
"""
Note:
We used the binary datatype where:
Government = 0
Private = 1
"""
# Display the unique values and their counts to verify the transformation
display(df['insurance_type'].value_counts())
```

```python
df['smoker'] = df['smoker'].replace({'No': 0, 'Yes': 1})
"""
Note:
We used the binary datatype where:
No = 0
Yes = 1
"""
# Display the unique values and their counts to verify the transformation
display(df['smoker'].value_counts())
```

Split data into tree parts ( Training , Valdiation , Testing ).

```python
x, x_test, y, y_test = train_test_split(data_input, data_output, test_size = 0.2 , random_state = 1)
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size = 0.33 , random_state = 1)
```

```python
#Import Libraries
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
#make a dictionary with the models to simplify model comparison
models = {
    "Linear Regression": LinearRegression(),
    "SVR": SVR(),
    "Decision Tree Regressor": DecisionTreeRegressor(),
    "KNN Regressor": KNeighborsRegressor(),
    "Random Forest Regressor": RandomForestRegressor()

}
```

```python
#compare models based on R2 score and RMSE and then choose best one to store
result = []
for name , model in models.items():
    model.fit(x_train , y_train)
    y_pred = model.predict(x_test)
    r2 = r2_score(y_test , y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))

    result.append({
        'Model' : name,
        'R2_Score': r2,
        'RMSE': rmse
    })

results = pd.DataFrame(result)
results
```
